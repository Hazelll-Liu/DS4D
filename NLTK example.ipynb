```python
import os

#得到所有文件名
def ref(path):
    list_all=[]
    files = os.listdir(path)
    for filename in files:
        # 拼接文件路径
        path1 = path+'/'+filename
        # 遍历叶节点文件夹
        list_all.append(path1)
    return list_all

#得到文件内容
def read_file(path):
    list_info=[]
    with open (path,encoding='utf-8') as f:
        list_info.append(f.read())
    return list_info
```

```python
ref('data20201101')
```

```python
#得到文件所有内容
all_content=[]
for i in ref('data20201102'):
    if i !=' ':
        all_content.append(read_file(i)[0])
all_content
```

```python
llll=[]
for i in range(0,40):
    llll.append(all_content[i])
```

```python
#进行分词
##[[],[],[]]
import nltk
list_clean_info1=[]
list_clean_info2=[]
for i in all_content:
    list_clean_info1.append(nltk.word_tokenize(i))
for j in list_clean_info1:
    if len(j)!=0:
        for k in j:
            list_clean_info2.append(k)
```

```python
english_punctuations = [',', '.', ':', ';', '?', '(', ')', '[', ']', '&', '!', '*', '@', '#', '$', '%','||','|','``','-',"''",'--',"'"]
clean_info=[]
for k in list_clean_info2:
    if k not in english_punctuations:
        clean_info.append(k)
clean_info
```

```python
#标注词性 并获得所有名词
from nltk.tag import pos_tag
from nltk.tokenize import word_tokenize
word_tag=pos_tag(clean_info)
all_words = []
for i in word_tag:
    if i[1]=='NN':
        all_words.append(i[0])
all_words
```

```python
#统计词频
freq ={}
for word in all_words:
    if word in freq:
        freq[word] += 1
    else:
        freq[word] = 1
freq
```

```python
top_freq=list(freq.items())[:8]
```

```python
top_freq
```

```python
#构建dataFrame的列表
list_wods = []
list_nuim = []
for i in top_freq:
    list_wods.append(i[0])
for j in top_freq:
    list_nuim.append(j[1])
```

```python
import seaborn as sns
import pandas as pd
df = pd.DataFrame()
df['word']=list_wods
df['num']=list_nuim
```
